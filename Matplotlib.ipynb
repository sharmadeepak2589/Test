```python
fig,ax = plt.subplots()
ax.plot(df_date.index,df_date["co2"])
ax.plot(df_date.index,df_date["relative_temp"])
plt.show()
```

```python
fig,ax = plt.subplots()   # two graphs with two different y axis with help of twinx
ax.plot(df_date.index,df_date["co2"])
ax2 = ax.twinx()
ax2.plot(df_date.index,df_date["relative_temp"])
plt.show()
```

```python
fig,ax = plt.subplots()   # two graphs showing different color
ax.plot(df_date.index,df_date["co2"],color="g",label="CO2")
ax.set_ylabel("CO2",color="g")
ax.tick_params("y",colors="g")
ax2 = ax.twinx()
ax2.plot(df_date.index,df_date["relative_temp"],color="indianred",label="Relative Temp.")
ax2.set_ylabel("Relative Temprature",color="indianred")
ax2.tick_params("y",colors="indianred")
plt.legend()
plt.show()
```

# Bar Charts

```python
fig,ax = plt.subplots(1,3,figsize=(20,3))
ax[0].bar(df_medals.index,df_medals["Gold"],color="g")
ax[0].set_xticklabels(df_medals.index,rotation=45)
ax[0].set_xlabel("Counrty")
ax[0].set_ylabel("Number of medals")


ax[1].bar(df_medals.index,df_medals["Silver"],color="b")
ax[1].set_xticklabels(df_medals.index,rotation=45)
ax[1].set_xlabel("Counrty")

ax[2].bar(df_medals.index,df_medals["Bronze"],color="orange")
ax[2].set_xticklabels(df_medals.index,rotation=45)
ax[2].set_xlabel("Counrty")
plt.show()
```

# Stacked Bar chart

```python
fig,ax = plt.subplots()
ax.bar(df_medals.index,df_medals["Gold"],color="g",label="Gold")
ax.bar(df_medals.index,df_medals["Silver"],color="b",label="Silver",bottom=df_medals["Gold"])
ax.bar(df_medals.index,df_medals["Bronze"],color="orange",label="Bronze",bottom=df_medals["Gold"]+df_medals["Silver"])
ax.set_xticklabels(df_medals.index,rotation=45)
ax.set_xlabel("Counrty")
plt.legend()
ax.set_ylabel("Number of medals")
```

```python
fig,ax = plt.subplots(1,3,figsize=(20,3))
ax[0].barh(df_medals.index,df_medals["Gold"],color="g")
ax[0].set_xticklabels(df_medals["Gold"])
ax[0].set_xlabel("Counrty")
ax[0].set_ylabel("Number of medals")


ax[1].barh(df_medals.index,df_medals["Silver"],color="b")
ax[1].set_xticklabels(df_medals["Silver"])
ax[1].set_xlabel("Counrty")

ax[2].barh(df_medals.index,df_medals["Bronze"],color="orange")
ax[2].set_xticklabels(df_medals["Bronze"])
ax[2].set_xlabel("Counrty")
plt.show()
```
